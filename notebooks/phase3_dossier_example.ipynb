# Phase 3 — Dossier Example

**Purpose:** Demonstrate `build_dossier()` for a single symbol, render to Markdown, and show `to_dict()` output.
Uses the synthetic market fixture; no live data required.

```python
# Cell 1: Imports and setup
import json
import pathlib
import sys
from datetime import date

import ah_research
from ah_research.analysis import Dossier, build_dossier

sys.path.insert(0, str(pathlib.Path().resolve() / "tests"))
from fixtures.phase2.synthetic_market import build_synthetic_market

code_version = getattr(ah_research, "__version__", "dev")
print(f"Python : {sys.version.split()[0]}")
print(f"ah-research: {code_version}")
```

```python
# Cell 2: Build synthetic market with 10 years of history
SYMBOL = "600000.SH"
ASOF = date(2024, 12, 31)

repo = build_synthetic_market(
    start=date(2014, 1, 1),
    end=ASOF,
    symbols=[SYMBOL],
)

print(f"Symbol  : {SYMBOL}")
print(f"As-of   : {ASOF}")
print(f"History : 2014-01-01 to {ASOF}  (~10 years synthetic data)")
```

```python
# Cell 3: Build dossier
dossier: Dossier = build_dossier(SYMBOL, repo, asof=ASOF)

print(f"Dossier built for   : {dossier.symbol}")
print(f"As-of               : {dossier.asof}")
print(f"Sector              : {dossier.overview.sector_l1}")
print(f"Fundamentals years  : {len(dossier.fundamentals.revenue_series)}")
print(f"OE latest FY        : {dossier.owner_earnings.latest_fy:.2f}")
print(f"PE current          : {dossier.valuation_bands.pe_current:.2f}")
print(f"PE percentile       : {dossier.valuation_bands.pe_current_percentile:.1f}%")
print(f"Dividend grade      : {dossier.dividend_history.consistency_grade}")
print(f"AH premium section  : {'present' if dossier.ah_premium else 'not applicable'}")
print(f"Warnings            : {dossier.metadata.warnings or 'none'}")
```

```python
# Cell 4: Render to Markdown (English)
md_en = dossier.to_markdown(language="en")
print(md_en[:2000])
if len(md_en) > 2000:
    print(f"... ({len(md_en) - 2000} more chars truncated)")
```

```python
# Cell 5: Render to Markdown (Chinese)
md_zh = dossier.to_markdown(language="zh")
print(md_zh[:2000])
if len(md_zh) > 2000:
    print(f"... ({len(md_zh) - 2000} more chars truncated)")
```

```python
# Cell 6: to_dict() round-trip
as_dict = dossier.to_dict()
blob = json.dumps(as_dict, default=str)
print(f"JSON bytes          : {len(blob)}")
print()
# Show top-level keys
print("Top-level keys:", list(as_dict.keys()))

# Show PE bands
vb = as_dict.get("valuation_bands", {})
if vb:
    print()
    print("PE bands:", vb.get("pe_bands"))
```
