# Phase 4.4 — Screener Filings Enrichment Example

Demonstrates `enrich_with_filings()`: add 5 filing-derived columns to a
symbol-indexed DataFrame.

````python
import pandas as pd

from ah_research.filings import enrich_with_filings
````

````python
# Small demo DataFrame — one real symbol, one fake
df = pd.DataFrame({"price": [1234.56, 99.99]}, index=["600519.SH", "999999.SH"])
df.index.name = "symbol"
df
````

````python
out = enrich_with_filings(df)
print(out.columns.tolist())
out
````

````python
# Correctness assertion
assert out.shape[1] >= df.shape[1] + 5, (
    f"Expected at least {df.shape[1] + 5} columns, got {out.shape[1]}"
)
# New columns present
for col in ("has_ipo", "n_annual", "latest_annual_year", "n_research", "has_profile"):
    assert col in out.columns, f"Missing column: {col}"
# Input not mutated
assert list(df.columns) == ["price"]
print("All assertions passed.")
````
