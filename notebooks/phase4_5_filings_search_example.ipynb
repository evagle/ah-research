# Phase 4.5 — Filings Text Search Example

Demonstrates `FilingsRepository.search()` — grep across the local filings corpus.

Searches Moutai (600519.SH) for the term "护城河" (moat). Skips gracefully if data is absent.

```python
from pathlib import Path

from ah_research.filings import FilingsRepository

FILINGS_ROOT = Path("data/filings")
MOUTAI = "600519.SH"

repo = FilingsRepository(root=FILINGS_ROOT)
```

```python
# Skip gracefully if Moutai data is absent
if not (FILINGS_ROOT / MOUTAI).exists():
    print(f"[skip] No filings found at {FILINGS_ROOT / MOUTAI}")
    print("Download filings first: ah filings download 600519.SH")
else:
    hits = repo.search("护城河", symbols=[MOUTAI])
    print(f"Found {len(hits)} hits for '护城河' in {MOUTAI}")
    print()
    for hit in hits[:5]:
        f = hit.filing
        year_info = f"year={f.year}" if f.year else (f"date={f.date}" if f.date else "")
        print(f"--- {f.symbol} | {f.kind} {year_info} | line {hit.line_no} ---")
        print(hit.line)
        print()
```

```python
# Assertion: at least one hit if data is present
if (FILINGS_ROOT / MOUTAI).exists():
    assert len(hits) >= 1, f"Expected >=1 hits for 护城河 but got {len(hits)}"
    print("Assertion passed: found at least one hit.")
```

## Annual-only search

Filter to annual reports only and search for a common term.

```python
if (FILINGS_ROOT / MOUTAI).exists():
    annual_hits = repo.search("公司", symbols=[MOUTAI], kinds=["annual"], max_hits_per_file=3)
    print(f"Annual '公司' hits (max 3/file): {len(annual_hits)}")
    for h in annual_hits[:5]:
        print(f"  {h.filing.year} | line {h.line_no}: {h.line[:60]}")
```
