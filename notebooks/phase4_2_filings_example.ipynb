# Phase 4.2 — Filings + Profile Repositories Example


```python
from ah_research.filings import FilingsRepository, ProfileRepository
```

## Filings

```python
fr = FilingsRepository()
print("Symbols:", fr.list_symbols())
filings = fr.list_filings("600519.SH")
print(f"{len(filings)} filings for 600519.SH")
for f in filings[:6]:
    print(f"  {f.kind:8}  year={f.year}  title={f.title!r}  path={f.path.name}")
```

## Latest annual

```python
latest = fr.latest_annual("600519.SH")
print(f"Year: {latest.year}")
print(f"First 500 chars:\n{latest.text[:500]}")
```

## Profiles

```python
pr = ProfileRepository()
print("Profile symbols:", pr.list_symbols())
profile = pr.latest("600519.SH")
if profile is not None:
    print(f"Date: {profile.date}  n_sections: {len(profile.sections)}")
    for name in list(profile.sections)[:10]:
        print(f"  - {name}")
```

```python
assert fr.list_symbols(), "no tickers under data/filings/"
assert pr.list_symbols(), "no profiles under profiles/"
print("All checks passed")
```
