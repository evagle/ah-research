# Phase 4.6 — Corpus Summary Example

Demonstrates `build_corpus_summary()` — one-glance audit of local research coverage.

Uses the Phase 4.2 fixture data so the notebook runs headlessly without real downloads.

```
from pathlib import Path

from ah_research.filings import FilingsRepository, ProfileRepository, build_corpus_summary

# Point at the fixture directories so this runs headlessly in CI
_FIXTURE = Path("tests/fixtures/phase4_2")
FILINGS_ROOT = _FIXTURE / "filings"
PROFILES_ROOT = _FIXTURE / "profiles"

fr = FilingsRepository(root=FILINGS_ROOT)
pr = ProfileRepository(root=PROFILES_ROOT)
df = build_corpus_summary(fr, pr)
print(df.to_string())
```

```
# Verify shape and key column values
assert not df.empty, "Expected non-empty DataFrame"
assert list(df.columns) == [
    "symbol",
    "n_annual",
    "latest_annual_year",
    "has_ipo",
    "n_research",
    "latest_research_date",
    "has_profile",
    "latest_profile_date",
    "profile_age_days",
    "annual_staleness_years",
], f"Unexpected columns: {list(df.columns)}"
row = df[df["symbol"] == "600000.SH"].iloc[0]
assert row["n_annual"] == 2, f"Expected n_annual=2, got {row['n_annual']}"
assert row["has_ipo"], f"Expected has_ipo=True, got {row['has_ipo']}"
print("All assertions passed.")
```
